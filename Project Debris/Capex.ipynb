```python
import pandas as pd
file = 'Z:/My Drive/Deals/2. Client work/4. Project Debris/2. Execution/12. PPE/capex GL.xlsx'
df = pd.read_excel(file)
```

```python
new_df = df[6:]
```

```python
new_df.head()
```

```python
list_to_remove = ['Ledger Accounts', 'Date']

# loop through the rows of the dataframe
for index, row in new_df.iterrows():
    # check if the second column contains any of the values in list_to_remove
    if row[0] in list_to_remove:
        # remove the row
        new_df.drop(index, inplace=True)
        
new_df_without_ledger = new_df.reset_index(drop=True)
```

```python
# save the new dataframe to an Excel file
new_df_without_ledger.to_excel('C:/Users/kmeiermatt001/Desktop/Cleaned capex GL.xlsx', index=False)
```
