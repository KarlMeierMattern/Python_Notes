```python
import pandas as pd

file = 'C:/Users/kmeiermatt001/Desktop/Data/FY17-19 Ladybird Mgmt ACCs.xlsx'
excel_file = pd.ExcelFile(file)
data = pd.DataFrame()

# Loop through each sheet in the file and append the data to the main dataframe.
for sheet_name in excel_file.sheet_names:
    sheet_data = excel_file.parse(sheet_name)
    data = data.append(sheet_data, ignore_index=True)

# this will print the sheet names
```

```python
data.head(10)
```

```python
# Specify how="all" so that only rows where all values are NaN are dropped.
new_data = data.dropna(how="all")
```
